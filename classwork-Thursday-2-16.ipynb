```python
nature_birds = {}
#read in file; adding encoding = "ISO-8859.15" to solve utf-8 problem
ff = open("formatted-nature.csv","r",encoding = "ISO-8859.15")
ss = ff.readlines()[1:] #removes the header
for line in ss:
    sp = line.split(",")[0]
    fam = line.split(",")[2]
    nature_birds[sp]=fam
ff.close()
```

```python
def makeBirdDict(filename):
    # this function opens the data file from nature-birds and extracts the species and families as a dictionary
    BirdDict= {}
    fobj= open(filename, "r",encoding = "ISO-8859.15") # open the file
    all_records = fobj.readlines()[2:] # read the file (list)
    for line in all_records:
        record_elements = line.split(",") # split lines on comma (this produces another list)
        species = record_elements[0]
        family = record_elements[2]
        BirdDict[species] = family # stick species and family into birdDict

    return BirdDict
```

```python
thomas_birds = makeBirdDict("formatted-nature.csv")
species = list(thomas_birds.keys())
species.sort() # sorted(speceis) will not sort it forever, it will only show you the sorted species one time
### U can also do below 3 lines, to do it in 3 lines, rather than 1
    #first_species = species[0]
    #firsr_fam = thomas_birds[first_species]
    #print (first_species, firsr_fam)

print("Family of " + species[0] + " is " + thomas_birds[species[0]])
```
